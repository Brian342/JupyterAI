<h2><u>About Dataset</h2></u>
Mental health is an essential part of people’s lives and society. Poor mental health affects our well-being, our ability to work, and our relationships with friends, family, and community.

Mental health conditions are not uncommon. Hundreds of millions suffer from them yearly, and many more do over their lifetimes. It’s estimated that 1 in 3 women and 1 in 5 men will experience major depression in their lives. Other conditions, such as schizophrenia and bipolar disorder, are less common but still have a large impact on people’s lives.

<h3><u>Mental health data based on surveys</h3></u>
Aside from the formal diagnoses, mental illnesses can be measured using surveys and screening questionnaires.
These tend to ask people about symptoms similar to those in diagnostic manuals, but they can be used more easily and widely because the data does not need to be collected by a healthcare professional.

<h3><u>What are the strengths and limitations of survey data?</u></h3>
<h4>Strengths of survey data</h4>

<u><i>survey data on mental health has two major strengths.</i></u>

- First, it involves structured interviews – people are asked a consistent set of questions regardless of their interviewer. This helps to ensure the data is more comparable between interviewers and across time.12
- Second, surveys include a much wider range of people in the population, including those who would not seek treatment due to a lack of awareness, costs, or other concerns.

<h3><u>Limitations of survey data</h3></u>

<u> <i> There are also several limitations of survey data to keep in mind.</u></i>
- One is that people may not share their symptoms in surveys, as they may not feel comfortable sharing them with researchers. Along with this, people’s comfort in sharing mental health symptoms may vary between countries and over time.16
- Another limitation is that people may not remember their symptoms, especially when they are asked to recall symptoms in their lifetime so far. This can be challenging for older people whose symptoms may have occurred decades ago.

```python
# Dealing with outliners
def outliners (df):
    check_outliners = []
    for cols in df.select_dtypes(include='float64').columns:
        # compute Q!, Q2 and IQR
        Q1 = df[cols].quantile(0.25)
        Q3 = df[cols].quantile(0.75)
        IQR = Q3 -Q1

        # compute outliner boundaries
        lower_boundary = Q1 - 1.5 * IQR
        upper_boundary = Q3 + 1.5 * IQR

        # count outliner
        outliners = df[(df[cols]< lower_boundary) | (df[cols] > upper_boundary)]
        outliner_count = outliners.shape[0] # returns the columns number
        outliner_percentage = (outliner_count / df.shape[0]) * 100

        check_outliners.append({
            'columns': cols,
            'outliner_count': outliner_count,
            'outliner_percentage': outliner_percentage,
            'lower_boundary': lower_boundary,
            'upper_boundary': upper_boundary
        })
    outliner_df = pd.DataFrame(check_outliners)
    return outliner_df
        
```

```python
outliners(df)
```

```python
# removng the outliners on the dataset
def remove_outliners(df: pd.DataFrame):
    df_no_outliner = df.copy()
    for cols in df.select_dtypes(include='float64').columns:
        # Compute Q1, Q2 and IQR
        Q1 = df[cols].quantile(0.25)
        Q3 = df[cols].quantile(0.75)
        IQR = Q3 -Q1

        # compute outliner boundaries
        lower_boundary = Q1 - 1.5 * IQR
        upper_boundary = Q3 + 1.5 * IQR
        # replacing outliners with Nan values
        df_no_outliner[cols] = df[cols].apply(lambda x : np.nan if(pd.notna(x) and (x < lower_boundary or x > upper_boundary))
                                              else x)
    return df_no_outliner
```

```python
clean_data = remove_outliners(df)
clean_data.sample(10)
```

```python
# checking the percentage of the outliner
def missing_value_per(df: pd.DataFrame):
    missing_per = (df.isna().sum() / len(df)) * 100
    missing_df = pd.DataFrame({"columns": df.columns,"MissingPer (%)": missing_per})
    missing_df = missing_df[missing_df["MissingPer (%)"] > 0].sort_values(by="MissingPer (%)", ascending=False)
    return missing_df
```

```python
# Are there significant differences in prevalence between developed vs. developing countries?
developed_countries = ['United States', 'Germany', 'Japan', 'United Kingdom', 'France',
                       'Canada', 'Italy', 'Australia']
clean_data['Developement_status'] = clean_data['Entity'].apply(lambda x: 'Developed' if x in developed_countries else 'Developing')
disorder_status = clean_data.groupby('Developement_status')[['Schizophrenia disorders',	'Depressive disorders',	'Anxiety disorders',
                    'Bipolar disorders','Eating disorders']].mean().T
```

```python
from scipy.stats import ttest_ind

result = {}
for disorders in disorder_status.index:
    developed = clean_data[clean_data['Developement_status']=='Developed'][disorders]
    developing = clean_data[clean_data['Developement_status'] == 'Developing'][disorders]
    t_stat, p_value = ttest_ind(developed, developing, nan_policy='omit')
    result[disorders] = p_value

for disorder, p in result.items():
    print(f"{disorder}: p-value = {p:.4f} {'*' if p < 0.05 else ''}")
```

```python
# visualization
plt.figure(figsize=(8, 5))
sns.barplot(data=disorder_status.reset_index().melt(id_vars='index'),
           y='value',
           x='index',
           hue='Developement_status',
            orient='v',
           palette={'Developed': '#69d', 'Developing': '#ff7f0e'}
)
plt.title('Mental illiness prevalence: Developed vs Developing')
plt.ylabel('Disorders')
plt.xlabel('mean Prevelance %')
plt.xticks(rotation=25)
plt.legend(title='Status')
plt.show()
```

```python
# Filling missing values with median
for cols in merge.select_dtypes(include='float64'):
    merge[cols] = merge[cols].fillna(merge[cols].median())

# checking the percentage of the missing value
def missing_value_per(df: pd.DataFrame):
    missing_per = (df.isna().sum() / len(df)) * 100
    missing_df = pd.DataFrame({"columns": df.columns,"MissingPer (%)": missing_per})
    missing_df = missing_df[missing_df["MissingPer (%)"] > 0].sort_values(by="MissingPer (%)", ascending=False)
    return missing_df
# running the function 
missing_value_per(merge)
```

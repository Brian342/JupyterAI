# HR Employee Attrition
- <b><h4>Definition</h4></b> HR analytics is the process of collecting and analyzing Human Resource (HR) data in order to improve an organization’s workforce performance.
The process can also be referred to as talent analytics, people analytics, or even workforce analytics. This method of data analysis takes data that is routinely collected by HR and correlates it to HR and organizational objectives. Doing so provides measured evidence of how HR initiatives are contributing to the organization’s goals and strategies.

- <b><h4>Need of HR Analytics</h4></b> Most organizations already have data that is routinely collected, so why the need for a specialized form of analytics? Can HR not simply look at the data they already have? Unfortunately, raw data on its own cannot actually provide any useful insight. It would be like looking at a large spreadsheet full of numbers and words. Once organized, compared and analyzed, this raw data provides useful insight.They can help answer
- <i><b>questions like:</i></b>

1) What patterns can be revealed in employee turnover?
2) How long does it take to hire employees?
3) What amount of investment is needed to get employees up to a fully productive speed?
4) Which of our employees are most likely to leave within the year?
5) Are learning and development initiatives having an impact on employee performance?

- <b><h4>The process of HR Analytics</b></h4>

HR Analytics is made up of several components that feed into each other.

1) To gain the problem-solving insights that HR Analytics promises, data must first be collected.
2) The data then needs to be monitored and measured against other data, such as historical information, norms or averages.
3) This helps identify trends or patterns. It is at this point that the results can be analyzed at the analytical stage.
4) The final step is to apply insight to organizational decisions.

```python
# using a chi-square test to determine if there is a significant association between the frequency of promotions and attrition

from scipy.stats import chi2_contingency

# Create a contingency table
contingency_table = pd.crosstab(df_copy['YearsSinceLastPromotion'], df_copy['Attrition'])

# Perform chi-square test
chi2, p, dof, expected = chi2_contingency(contingency_table)
print(f"Chi-Square Statistic: {chi2}, p-value: {p}")
# interpretation: A low p-value(<0.05) indicates a significant association, between promotions and attritions
```

```python
# use a t-test to compare the mean salary hikes of employees who left and those who stayed
from scipy.stats import ttest_ind

# Split data into two groups
salary_hike_left = df_copy[df_copy['Attrition'] == 'Yes']['PercentSalaryHike']
salary_hike_stayed = df_copy[df_copy['Attrition'] == 'No']['PercentSalaryHike']

# Perform t-test
t_stat, p_value = ttest_ind(salary_hike_left, salary_hike_stayed)
print(f"T-Statistic: {t_stat}, p-value: {p_value}")
# A low p-value(<0.05) indicates a significant difference in salary hikes between employees who left and who stayed
```

```python
def check_outliers(df):
    """A function that checks outliers on the dataset"""
    outliner_data = []
    for cols in df.select_dtypes(include='int64').columns:
        # computer Q1, Q3, IQR
        q1 = df[cols].quantile(0.25)
        q3 = df[cols].quantile(0.75)
        IQR = q3 - q1

        # computer outliner boundaries
        lower_boundary = q1 - 1.5 * IQR
        upper_boundary = q3 + 1.5 * IQR

        # count outliners
        outliners = df[(df[cols] < lower_boundary) | (df[cols] > upper_boundary)]
        outliner_count = outliners.shape[0] # returns the columns number
        outliner_percentage = (outliner_count / df.shape[0]) * 100

        # Append data to the list
        outliner_data.append({
            'columns': cols,
            'outliner_count' : outliner_count,
            'outliner_percentage' : outliner_percentage,
            'lower_boundary' : lower_boundary,
            'upper_boundary' : upper_boundary
        })
        # convert list into a DataFrame
    outliner_df = pd.DataFrame(outliner_data)
    return outliner_df
```

```python
# drop outliners on the dataset
def drop_outliners(df: pd.DataFrame):
    # create a copy of the dataframe to avoid modification on the code
    df_no_outliners = df.copy()
    for cols in df.select_dtypes(include='int64').columns:
        # calculating the upper and lower quantile
        q1 = df[cols].quantile(0.25)
        q3 = df[cols].quantile(0.75)
        IQR = q3 - q1

        # calculate the quantiles
        lower_bound = q1 - 1.5 * IQR
        upper_bound = q3 + 1.5 * IQR

        # Replace outliners with NaN values 
        df_no_outliners[cols] = df[cols].apply(lambda x : np.nan if (pd.notna(x) and (x < lower_bound or x > upper_bound)) else x
                                              )
    return df_no_outliners
```

```python
# calculating missing values on the dataset
def missing_values_percentage(df: pd.DataFrame):
    missing_percentage = (df.isna().sum() / len(df)) * 100
    missing_df = pd.DataFrame({'columns': df.columns, 'missing (%)': missing_percentage})
    missing_df = missing_df[missing_df['missing (%)'] > 0].sort_values(by='missing (%)', ascending=False)
    
    return missing_df
```

```python
# filling missing values with median values
def fill_na(df):
    """Function to fillna values with median values
    Parameters:
    Takes in DataFrame df 
    returns: 
    A function that fills the null values with median values and makes the data change real time"""
    return df.fillna(df.median(), inplace=True)
```

```python
# Encoding categorical values
from sklearn.preprocessing import LabelEncoder

def encode_categorical_columns(df: pd.DataFrame, encoding_type: str = 'label') -> pd.DataFrame:
    """
    Encodes categorical columns in a DataFrame.

    Parameters:
        df (pd.DataFrame): The input DataFrame with categorical columns.
        encoding_type (str): The encoding type, either 'label' or 'onehot'.

    Returns:
        pd.DataFrame: A DataFrame with encoded categorical columns.
    """

    df_encoded = df.copy()

    categorical_columns = df_encoded.select_dtypes(include=['category']).columns

    if encoding_type == 'label':
        for col in categorical_columns:
            le = LabelEncoder()
            df_encoded[col] = le.fit_transform(df_encoded[col])

    elif encoding_type == 'onehot':
        df_encoded = pd.get_dummies(df_encoded, columns=categorical_columns, drop_first=True)
    else:
        raise ValueError("Unsupported encoding_type. Use 'label' or 'onehot' .")


    return df_encoded
```

```python
df_encoded = encode_categorical_columns(clean_data, encoding_type='label')
df_encoded.sample(10)
```

```python
# scaling the values
from sklearn.preprocessing import StandardScaler, MinMaxScaler
def scale_data(df: pd.DataFrame, target_column : str, scaling_type : str = 'standard') -> pd.DataFrame:
    """
    Scales numerical columns on the dataset
    
    parameters:
    df (pd.DataFrame): input the DataFrame
    scaling_type: returns the scale type in either standard or minmax
    target_column: the column most important
    
    Returns:
    returns a scaled dataset"""

    df_scaled = df.copy()

    numerical_columns = df_scaled.select_dtypes(include=(['int64', 'float64'])).columns
    columns_scaling = numerical_columns.difference([target_column])

    if scaling_type == 'standard':
        scaler = StandardScaler()
    elif scaling_type == 'minmax':
        scaler = MinMaxScaler()
    else:
        raise ValueError("Unsupported scaler, use standard or minmax")

    df_scaled[columns_scaling] = scaler.fit_transform(df_scaled[columns_scaling])

    return df_scaled
```

<h2>Checking for class imbalance</h2>

```python
# checking for class imbalance
from imblearn.over_sampling import SMOTE
def balanceWithSmote(df: pd.DataFrame, target_column: str):
    """Function used to check and balance the dataset
    parameters:
    df (pd.DataFrame) used to take a dataframe (Encoded or scaled) and
    a target column of type string
    
    Returns:
    X_resampled: (pd.DataFrame) resampled feature set
    Y_resample: (pd.Series) Resampled target column
    """
    X =df.drop(columns=[target_column])
    Y = df[target_column]

    smote = SMOTE(random_state=142)
    X_resampled, Y_resampled = smote.fit_resample(X, Y)

    # Convert back to dataframe
    X_resampled = pd.DataFrame(X_resampled, columns=X.columns)
    Y_resampled = pd.Series(Y_resampled, name=target_column)

    return X_resampled, Y_resampled
    
```

```python
# Balancing the DataSet
X_balanced, y_balanced = balanceWithSmote(df = df_scaled, target_column='Attrition')
```
